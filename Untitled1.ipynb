```python
class A:
    def hello(self):
        print("1",end=",")
    def hello(self):
        print("2",end=",")
class B(A):
    def __init__(self):
        self.hello()
    def hello(self):
        print("3",end=",")
        super().hello()
b=B()
b.hello()
```

```python
class A:
    money=5000
    def __init__(self):
        self.money+=A.money
class B(A):
    def display(self):
        self.money+=2000
        print(self.money)
obj=B()
obj.display()
```

```python
class A:
    money=100000
    def buy(self):
        super().buy()
        self.money=5000
class B(A):
    def buy(self):
        self.money=3000
class C(A):
    money=30000
    def buy(self):
        self.money=10000
class D(B,C):
    B.money+=C.money
    def buy(self):
        return self.money
obj=D()
print(obj.buy())
```

```python
from abc import ABC,abstractmethod
class Father(ABC):
    def display(self):
        pass
    def play(self):
        print('Abstract class')
class Son(Father):
    def play(self):
        print('Child class')
A=Son()
A.play()
```
